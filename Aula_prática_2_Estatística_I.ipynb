```python
# Gerando uma amostra de valores aleatÃ³rios para nossa distribuiÃ§Ã£o uniforme

import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import uniform
import statistics as st
import numpy as np


a = -1
b = 1
amostra = 1000
x_uniforme = np.random.uniform(a, b, amostra)
plt.figure(figsize=(10,6))
plt.subplot(211)
plt.plot(x_uniforme, color = 'black')
plt.xlabel('eixo horizontal')
plt.ylabel('NÃºmeros aleatÃ³rios entre -1 e 1')

plt.subplot(212)
plt.hist(x_uniforme, bins=5)
plt.xlabel('Classes')
plt.ylabel('FrequÃªncia')
```

```python
# FunÃ§Ã£o densidade de probabilidade

x1 = np.linspace(-2, 2, 100)
y1 = uniform.pdf(x1)
plt.figure(figsize=(10,6))
plt.plot(x1, y1)
```

```python
# FunÃ§Ã£o distribuiÃ§Ã£o acumulada

y = uniform.cdf(x1)
plt.figure(figsize=(10,6))
plt.plot(x1,y)
```

```python
# MÃ©dia da distribuiÃ§Ã£o

med_dados = st.mean(x_uniforme)
med_teorica = (a + b)/2

# VariÃ¢ncia

var_dados = st.pvariance(x_uniforme)
var_teorica = ((b-a)**2)/12

# Calculando uma funÃ§Ã£o dos dados

a_cal = med_dados*2 -( med_dados*2+np.sqrt(var_dados*12)/2)
b_cal = a_cal +np.sqrt(var_dados*12)

print(med_dados, med_teorica)
print(var_dados, var_teorica)
print(a_cal,b_cal)
```
